# Feature scoring overtime

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from ema_workbench import ema_logging, load_results
from ema_workbench.analysis import (get_ex_feature_scores,
                                    RuleInductionType)

ema_logging.log_to_stderr(level=ema_logging.INFO)


fn =  './data/data_kde7.tar.gz'

x, outcomes = load_results(fn)
#x = x.drop(['model', 'policy'], axis=1)
```

```python
#for outcome biomass


y = outcomes['Biomass mesopelagic fish'] 


all_scores = []

top_5 = set()
for i in range(2, y.shape[1], 8):
    data = y[:, i]
    scores = get_ex_feature_scores(x, data,
                                   mode=RuleInductionType.REGRESSION)[0]
    
    top_5 |= set(scores.nlargest(3, 1).index.values)   
    scores = scores.rename(columns={1:outcomes['TIME'][0, i]})
    all_scores.append(scores)
    
all_scores = pd.concat(all_scores, axis=1, sort=False)
all_scores = all_scores.loc[top_5, :]

scores=all_scores
title='biomass mesopelagic fish'
sns.heatmap(scores, cmap='viridis')
fig = plt.gcf()
ax = fig.get_axes()

fig.set_size_inches(15,5)
fig.suptitle('Feature scoring for uncertainties most influential for '+title)
shorttitle = title.replace(" ","")
plt.savefig('./Figures/feature_bm3.png')
plt.show()
```
