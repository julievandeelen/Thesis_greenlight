# Plotting results from Step 2

## Plot trade-offs & Policy over time

```python
from ema_workbench import ema_logging, load_results
from ema_workbench.analysis import parcoords
import matplotlib.pyplot as plt
import pandas as pd
from ema_workbench.analysis import parcoords
```

```python
results=pd.read_excel('./Data/results_convergence7_supercomp4.xlsx')
# drop unnamed column
results = results.drop(['Unnamed: 0'], axis=1)

results
```

```python
results2=results.T
results2 = results2.iloc[:-4]
fig=results2.plot(legend=False)
fig.set_xlabel("Time (2010-2100)")
fig.set_ylabel("Harvesting Quota in (Gt fish)")
plt.tick_params(
    axis='x',         
    which='both',    
    bottom=False,
    top=False,      
    labelbottom=False)
plt.savefig('./Figures/policy_sup4.png')

```

```python


data = results.loc[:, ['Average food provision by MF', 'Average vertical migration', 'Biomass MF 10th percentile', 'Final atmospheric C level', 'Inertia']]

limits = parcoords.get_limits(data)
paraxes = parcoords.ParallelAxes(limits, fontsize=8)
paraxes.plot(data)
paraxes.invert_axis('Final atmospheric C level')
fig=plt.gcf()
fig.set_size_inches(6,10)
plt.savefig('./Figures/tradeoffs_st2_sup4.png')
plt.show()
```
