```python
import matplotlib.pyplot as plt

from ema_workbench import ema_logging, load_results
from ema_workbench.analysis.plotting import lines, Density
import seaborn as sns

ema_logging.log_to_stderr(ema_logging.INFO)
```

```python
file_name = './data/data_kde100.tar.gz' 

experiments, outcomes = load_results(file_name)
```

```python
lines(experiments, outcomes,outcomes_to_show = 'Total vertical migration',density = Density.KDE)
fig = plt.gcf()

fig.set_size_inches(10,5)
ax = fig.get_axes()

ax[0].set_ylabel('GtC/Year')
for line in ax[0].get_lines():
    line.set_alpha(0.1)
sns.despine()
plt.savefig('Figures/lines_vm.png')
plt.show()
```

```python
lines(experiments, outcomes,outcomes_to_show = 'Biomass mesopelagic fish' ,density = Density.KDE)
fig = plt.gcf()

fig.set_size_inches(10,5)
ax = fig.get_axes()

ax[0].set_ylabel('Gt fish')
for line in ax[0].get_lines():
    line.set_alpha(0.1)
sns.despine()

plt.savefig('Figures/lines_bio.png')
plt.show()
```

```python
#
lines(experiments, outcomes,outcomes_to_show = 'Atmospheric C' ,density = Density.KDE)
fig = plt.gcf()

fig.set_size_inches(10,5)
ax = fig.get_axes()

ax[0].set_ylabel('GtC')
for line in ax[0].get_lines():
    line.set_alpha(0.1)
sns.despine()


plt.savefig('Figures/lines_atm.png')
plt.show()
```

```python
# subplot


lines(experiments, outcomes,outcomes_to_show = ['Total vertical migration', 'Atmospheric C', 'Food provision by MF', 'Biomass mesopelagic fish'] , group_by = 'policy' ,density = Density.KDE)
fig = plt.gcf()
fig.set_size_inches(13,15)
ax = fig.get_axes()

ax[0].set_ylabel('GtC/Year')
ax[2].set_ylabel('GtC')
ax[4].set_ylabel('Gt food/Year')
ax[6].set_ylabel('GtMF')

for line in ax[0].get_lines():
    line.set_alpha(0.05)
for line in ax[2].get_lines():
    line.set_alpha(0.05)
for line in ax[4].get_lines():
    line.set_alpha(0.05)
for line in ax[6].get_lines():
    line.set_alpha(0.05)
    
for i in [0,2,4,6]:
    ax[i].set(ylim=(0, None), xlim=(2010,2100))
    
sns.despine()
plt.savefig('Figures/lines_all_3pol.png',bbox_inches="tight")
plt.show()
```
